```python
X=cleaned_df.drop('concrete_compressive_strength',axis=1).values
y=cleaned_df['concrete_compressive_strength'].values
```

```python
X  ### .values convert data frame into array
```

```python
# When we use deep learning we must scale our data
# Before that we split our data into train and test part
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
#data scaling
scale=MinMaxScaler()   #scale on basis of median
X_train_scale=scale.fit_transform(X_train)
X_test_scale=scale.fit_transform(X_test)
```

```python
# Now number of neuron and no of hidden layer are done my experiment for our case we used as shown below
model.add(Dense(8,activation='relu'))
model.add(Dense(800,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(600,activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(400,activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(200,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(100,activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(1))
op=Adam(0.0000506)
model.compile(optimizer=op,loss='mse')
#Since it is regression we need to minimise mse 
```

```python
model.fit(X_train_scale,y_train,batch_size=128,epochs=400,validation_data=(X_test_scale,y_test))
```

```python
df_hist=pd.DataFrame(model.history.history)
```

```python
df_hist.plot()
```

```python
# clearly we can see the loss and val_loss(loss in testing set) is getting reduced so we can go for more epochs
```

```python
model.evaluate(X_test_scale,y_test)**0.5
```

```python
y_pred=model.predict(X_test_scale)
y_pred=y_pred.reshape(287,)
```
